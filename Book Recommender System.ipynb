```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
# scipy :- scientific python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
# use sparse matrix due to lot of zeros in book_pivot
# sparse matrix considered only non-zero values for calculating distances.
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
# brute means find distances from each and every.
# if we not put brute then they skip calculating some distances
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions=model.kneighbors(book_pivot.iloc[254, :].values.reshape(1,-1), n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
def recommend_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distances, suggestions=model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1,-1), n_neighbors=6)
    
    for i in range(len(suggestions)):
        if i==0:
            print("The suggestions for", book_name,"are :")
        if not i:
            print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('High Fidelity')
```
